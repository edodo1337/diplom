<h1>Расчеты КПД</h1>

```python
from iapws import IAPWS97, IAPWS95

def eff_func(row, pump_num):
    p1 = df[f'10LAB{pump_num}CP002_XQ01'][row]
    p2 = df[f'10LAB{pump_num}CP003_XQ01'][row]
    G = df[f'10LAB{pump_num}CF001_XQ01'][row]
    t = df["10LAB00FT902_XQ01"][row]
    I = df[f"10LAC{pump_num}CE021_XQ01"][row]
    U = df["10BBA00CE010_XQ01"][row]
    
    if all([(p2-p1)*G,U*I, t, p2, p1]):
        try:
            rho = IAPWS97(P=(p1+p2)/2, T=t+273).rho
        except:
            print(p1,p2,G,t, pump_num)
            raise Exception('here')
        ef = (p2-p1)*1000*G/(U*I*rho)/np.sqrt(3)
        ef = 0 if ef < 0 or ef >= 1 else ef
        
        return ef / np.sqrt(3)
        
    else:
        return 0


pump_1_eff = df.apply(lambda x: eff_func(x.name, 11), axis=1)
pump_2_eff = df.apply(lambda x: eff_func(x.name, 12), axis=1)
pump_3_eff = df.apply(lambda x: eff_func(x.name, 13), axis=1)
pump_4_eff = df.apply(lambda x: eff_func(x.name, 14), axis=1)
pump_5_eff = df.apply(lambda x: eff_func(x.name, 15), axis=1)

pump_eff_data = {
    'pump_1_eff':pump_1_eff,
    'pump_2_eff':pump_2_eff,
    'pump_3_eff':pump_3_eff,
    'pump_4_eff':pump_4_eff,
    'pump_5_eff':pump_5_eff,
}

```

<h3>Обучение с применением функции ReLU</h3>

```python
import sys
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras import optimizers

start_time = time()

class NNTrainer:
    def __init__(self, title, EPOCHS=3500, patience=300):
        self.EPOCHS = EPOCHS
        self.patience = patience
        self.title = title
    
    def build_model(self, train_dataset):
        pass
        
    def build_and_train(self, pump_num):
        pump_filtred_df = pd.DataFrame({
            'pump_in': filtred_data[f'pump_{pump_num}_in'],
            'pump_out': filtred_data[f'pump_{pump_num}_out'],
            'pump_mass': filtred_data[f'pump_{pump_num}_mass'],
            'pump_temp': filtred_data[f'pump_{pump_num}_temp'],
            'pump_revs': filtred_data[f'pump_{pump_num}_revs'],
            'pump_amper': filtred_data[f'pump_{pump_num}_amper'],
            'pump_eff': filtred_data[f'pump_{pump_num}_eff'],
        })

        train_dataset = pump_filtred_df.sample(frac=0.8,random_state=0)
        test_dataset = pump_filtred_df.drop(train_dataset.index)

        train_stats = train_dataset.describe()
        train_stats.pop("pump_eff")
        train_stats = train_stats.transpose()

        train_labels = train_dataset.pop('pump_eff')
        test_labels = test_dataset.pop('pump_eff')

        def norm(x):
            return (x - train_stats['mean']) / train_stats['std']

        normed_train_data = norm(train_dataset)
        normed_test_data = norm(test_dataset)
        
        EPOCHS = self.EPOCHS
        patience = self.patience

        model = self.build_model(train_dataset)

        early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=patience)
        
        title = self.title
        
        class PrintDot(keras.callbacks.Callback):
            def on_epoch_end(self, epoch, logs):
                if epoch%100== 0:
                    print(f"\rProgress {title} (#{pump_num}): {epoch}/{EPOCHS} mae:{logs['mae']}, mse:{logs['mse']}", end='')

        history = model.fit(normed_train_data, train_labels, epochs=EPOCHS,
                            validation_split = 0.2, verbose=0, callbacks=[early_stop, PrintDot()])
        loss, mae, mse = model.evaluate(normed_test_data, test_labels, verbose=2)
        print("Testing set (#{}) MAE: {:5.10f}, MSE: {:5.10f}".format(pump_num, mae, mse))
        test_predictions = model.predict(normed_test_data).flatten()
        return history.history, history.epoch, mae, mse, test_labels, test_predictions
        

class ReLU64_64(NNTrainer):
    def build_model(self, train_dataset):
        model = keras.Sequential([
        layers.Dense(64, activation='relu', input_shape=[len(train_dataset.keys())]),
        layers.Dense(64, activation='relu'),
        layers.Dense(1)
        ])

        optimizer = tf.keras.optimizers.RMSprop(0.001)

        model.compile(loss='mse',
                    optimizer=optimizer,
                    metrics=['mae', 'mse'])
        return model

    
class Sigmoid(NNTrainer):
    def build_model(self, train_dataset):
        model = keras.Sequential([
            layers.Dense(64, activation='sigmoid', input_shape=[len(train_dataset.keys())]),
            layers.Dense(64, activation='sigmoid'),
            layers.Dense(1)
        ])

        optimizer = tf.keras.optimizers.RMSprop(0.001)

        model.compile(loss='mse',
                    optimizer=optimizer,
                    metrics=['mae', 'mse'])
        return model
    
class Tanh(NNTrainer):
    def build_model(self, train_dataset):
        model = keras.Sequential([
            layers.Dense(64, activation='tanh', input_shape=[len(train_dataset.keys())]),
            layers.Dense(64, activation='tanh'),
            layers.Dense(1)
        ])

        optimizer = tf.keras.optimizers.RMSprop(0.001)

        model.compile(loss='mse',
                    optimizer=optimizer,
                    metrics=['mae', 'mse'])
        return model
    
class LeakyReLUTrainer(NNTrainer):
    def build_model(self, train_dataset):
        model = keras.Sequential([
        LeakyReLU(alpha=0.1, input_shape=[len(train_dataset.keys())]),
        LeakyReLU(alpha=0.1),
        layers.Dense(1)
        ])

        optimizer = tf.keras.optimizers.RMSprop(0.001)

        model.compile(loss='mse',
                    optimizer=optimizer,
                    metrics=['mae', 'mse'])
        return model
    
class Softmax(NNTrainer):
    def build_model(self, train_dataset):
        model = keras.Sequential([
        layers.Dense(64, activation='softmax', input_shape=[len(train_dataset.keys())]),
        layers.Dense(64, activation='softmax'),
        layers.Dense(1)
        ])

        optimizer = tf.keras.optimizers.RMSprop(0.001)

        model.compile(loss='mse',
                    optimizer=optimizer,
                    metrics=['mae', 'mse'])
        return model
    
class SELU(NNTrainer):
    def build_model(self, train_dataset):
        model = keras.Sequential([
        layers.Dense(64, activation='selu', input_shape=[len(train_dataset.keys())]),
        layers.Dense(64, activation='selu'),
        layers.Dense(1)
        ])

        optimizer = tf.keras.optimizers.RMSprop(0.001)

        model.compile(loss='mse',
                    optimizer=optimizer,
                    metrics=['mae', 'mse'])
        return model

class ELU(NNTrainer):
    def build_model(self, train_dataset):
        model = keras.Sequential([
        layers.Dense(64, activation='elu', input_shape=[len(train_dataset.keys())]),
        layers.Dense(64, activation='elu'),
        layers.Dense(1)
        ])

        optimizer = tf.keras.optimizers.RMSprop(0.001)

        model.compile(loss='mse',
                    optimizer=optimizer,
                    metrics=['mae', 'mse'])
        return model
    
class ReLU32_32(NNTrainer):
    def build_model(self, train_dataset):
        model = keras.Sequential([
        layers.Dense(32, activation='relu', input_shape=[len(train_dataset.keys())]),
        layers.Dense(32, activation='relu'),
        layers.Dense(1)
        ])

        optimizer = tf.keras.optimizers.RMSprop(0.001)

        model.compile(loss='mse',
                    optimizer=optimizer,
                    metrics=['mae', 'mse'])
        return model
    
class ReLU12_12_12(NNTrainer):
    def build_model(self, train_dataset):
        model = keras.Sequential([
        layers.Dense(12, activation='relu', input_shape=[len(train_dataset.keys())]),
        layers.Dense(12, activation='relu'),
        layers.Dense(12, activation='relu'),
        layers.Dense(1)
        ])

        optimizer = tf.keras.optimizers.RMSprop(0.001)

        model.compile(loss='mse',
                    optimizer=optimizer,
                    metrics=['mae', 'mse'])
        return model
```

```python
from multiprocessing import Pool, Process
import os

def plot_history(history, epochs, pump_num, trainer):
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    ax2 = ax1.twinx()
    
    
    ax1.plot(epochs, history['mae'], color='blue',
           label = f'Ошибка по модулю (ПЭН-{pump_num})') 
    ax2.plot(epochs, history['mse'], color='orange',
           label= f'Среднеквадратичная ошибка (ПЭН-{pump_num})')
    ax1.set_title(trainer.title)
    
    h1, l1 = ax1.get_legend_handles_labels()
    h2, l2 = ax2.get_legend_handles_labels()
    ax1.legend(h1+h2, l1+l2, loc="upper right")
    
#     ax1.set_ylim([0, min(history['mae']*2)])
#     ax2.set_ylim([0, min(history['mse']*2)])
 
    plt.show()
    
    
    my_path = os.path.abspath(os.getcwd())
    directory=  f'{my_path}/{trainer.title}'
    if not os.path.exists(directory):
        os.makedirs(directory)    
    fig.savefig(f'{my_path}/{trainer.title}/history_#{pump_num}.png')
    


def plot_after_train(test_labels, test_predictions, history, epochs, pump_num, trainer):
    plot_history(history, epochs, pump_num, trainer)
    fig8 = plt.figure()
    ax8 = fig8.add_subplot(111)

    ax8.plot(test_predictions, color='blue', label='Предсказанное значение')
    ax8.plot(test_labels.reset_index()['pump_eff'], color='orange', label='Фактическое значение')
    ax8.set_ylabel(f'КПД ПЭН-{pump_num}')
    ax8.set_title(trainer.title)
    ax8.legend(loc='lower right')
    
    
    my_path = os.path.abspath(os.getcwd())
    directory=  f'{my_path}/{trainer.title}'
    if not os.path.exists(directory):
        os.makedirs(directory)
    
    fig8.savefig(f'{my_path}/{trainer.title}/testing#{pump_num}.png')
    
    plt.show()    
    
def calc_and_save(trainers, processes=8):
    pool = Pool(processes)
    
    out_dicts = []
    
    futures = []
    for trainer in trainers:
        for i in range(5):
            futures.append(pool.apply_async(trainer.build_and_train, (i+1,)))
    
    _results = [f.get() for f in futures]
        
#         results = list(pool.map_async(trainer.build_and_train, [1,2,3,4,5]))
    out_dict = {}
    counter = 0

    for results in _results:
        plot_after_train(results[4], results[5], results[0], results[1], i+1, trainer)
        out_dict[counter+1] = (results[0]['loss'][-1], results[0]['mae'][-1],results[0]['mse'][-1],
                         results[1][-1], results[2], results[3], counter)
        counter += 1
        
        if counter == 5:
            out_dicts.append(out_dict)
            out_dict = {}
            counter = 0
    
    pool.close()
    pool.join()
    
    return out_dicts
    
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

def build_and_train(pump_num, estimators=1000):
    pump_filtred_df = pd.DataFrame({
            'pump_in': filtred_data[f'pump_{pump_num}_in'],
            'pump_out': filtred_data[f'pump_{pump_num}_out'],
            'pump_mass': filtred_data[f'pump_{pump_num}_mass'],
            'pump_temp': filtred_data[f'pump_{pump_num}_temp'],
            'pump_revs': filtred_data[f'pump_{pump_num}_revs'],
            'pump_amper': filtred_data[f'pump_{pump_num}_amper'],
            'pump_eff': filtred_data[f'pump_{pump_num}_eff'],
        })

    train_dataset = pump_filtred_df.sample(frac=0.8,random_state=0)
    test_dataset = pump_filtred_df.drop(train_dataset.index)

    train_stats = train_dataset.describe()
    train_stats.pop("pump_eff")
    train_stats = train_stats.transpose()

    train_labels = train_dataset.pop('pump_eff')
    test_labels = test_dataset.pop('pump_eff')

    features = pump_filtred_df[['pump_in', 'pump_out', 'pump_mass', 'pump_temp', 'pump_revs', 'pump_amper']]
    labels = pump_filtred_df['pump_eff']
    train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)

    rf = RandomForestRegressor(n_estimators = estimators, random_state = 42)
    rf.fit(train_features, train_labels)    
    predictions = rf.predict(test_features)
    errors = abs(predictions - test_labels)
    mape = 100 * (errors / test_labels)
    accuracy = 100 - np.mean(mape)
    
    print(f"Accuracy #{pump_num} (estimators={estimators}): {accuracy}")
    print(f"Mean Absolute Error #{pump_num}: {round(np.mean(errors), 5)}")
     
```

```python
from multiprocessing import Pool, Process
import os

def plot_history(history, epochs, trainer):
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    ax2 = ax1.twinx()

    ax1.plot(epochs, history['mae'], color='orange',
           label = f'Ошибка по модулю') 
    ax2.plot(epochs, history['mse'], color='blue',
           label= f'Среднеквадратичная ошибка')
    ax1.set_title(trainer.title)
    
    ax1.set_ylim(0, min(history['mae'])*5)
    ax2.set_ylim(0, min(history['mse'])*5)
    
    h1, l1 = ax1.get_legend_handles_labels()
    h2, l2 = ax2.get_legend_handles_labels()
    ax1.legend(h1+h2, l1+l2, loc="upper right")

    
    plt.show()
        
    my_path = os.path.abspath(os.getcwd())
    directory=  f'{my_path}/{trainer.title}'
    if not os.path.exists(directory):
        os.makedirs(directory)
    
    fig.savefig(f'{my_path}/{trainer.title}/history_#all.png')
    


def plot_after_train(test_labels, test_predictions, history, epochs, trainer):
    plot_history(history, epochs, trainer)
    fig8 = plt.figure()
    ax8 = fig8.add_subplot(111)
    
    ax8.plot(test_predictions, color='blue', label='Предсказанное значение', linewidth=3)
    ax8.plot(test_labels.values, color='orange', label='Фактическое значение')
    ax8.set_ylabel(f'Температура пит. воды')
    ax8.set_title(trainer.title)
    ax8.legend(loc='lower right')
    
    ax8.set_ylim(test_labels.values.min() / 1.2, test_labels.values.max() * 1.2)
    
    my_path = os.path.abspath(os.getcwd())
    directory=  f'{my_path}/{trainer.title}'
    if not os.path.exists(directory):
        os.makedirs(directory)
    
    fig8.savefig(f'{my_path}/{trainer.title}/testing#all.png')
    
    plt.show()    
    
def calc_and_save(trainers, mode, processes=8):
    pool = Pool(processes)
    
    out_dicts = []
    
    futures = []
    for trainer in trainers:
        futures.append(pool.apply_async(trainer.build_and_train, (mode,)))
    
    _results = [f.get() for f in futures]
        
    out_dict = {}
    counter = 0

    for results, trainer in zip(_results, trainers):
        if mode == 'train':
            plot_after_train(results[4], results[5], results[0], results[1], trainer)
        out_dict[trainer.title] = (results, trainer)
        
    pool.close()
    pool.join()
    
    return out_dict
    
```
